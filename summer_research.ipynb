```python
import pandas as pd
import matplotlib.pyplot as plt

# loads mutation file
file_path = "data_mutations.txt"
mutation_data = pd.read_csv(file_path, sep='\t')

# preview first few rows
print(mutation_data.head())
print(mutation_data.columns)  #column names

#count mutations per gene
top_genes = mutation_data['Hugo_Symbol'].value_counts().head(10)

#prints top 10 mutated genes
print("Top 10 mutated genes:")
print(top_genes)


plt.figure(figsize=(10,6))
top_genes.plot(kind='bar', color='purple')
plt.title('Top 10 Mutated Genes in Medulloblastoma Group 4')
plt.xlabel('Gene')
plt.ylabel('Mutation Count')
plt.xticks(rotation=45)
plt.show()

```
